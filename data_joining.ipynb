```python
import pandas as pd

# Creating a DataFrame with product data
products = {
    "product_id": [1, 2, 3, 4],
    "product_name": ["Laptop", "Tablet", "Smartphone", "Smartwatch"],
    "category": ["Electronics", "Electronics", "Electronics", "Wearables"],
    "price": [1200, 300, 800, 250],
}

df_products = pd.DataFrame(products)
print(df_products)

# Creating a DataFrame with order details
order_details = {
    "order_id": [101, 102, 103, 104, 105, 106, 107],
    "product_id": [2, 2, 1, 2, 4, 1, 5],
    "customer_id": [201, 202, 202, 204, 203, 201, 202],
    "quantity": [1, 2, 1, 3, 2, 2, 4],
}

df_order_details = pd.DataFrame(order_details)
print(df_order_details)
```

## Inner join of products and orders

_Which products have been ordered?_
- Laptop, Tablet and Smartwatch have been ordered.

```python
pd.merge(df_products, df_order_details, on="product_id", how="inner")
```

## Left join of products and orders

_Which products haven't been ordered?_
- Smartphone hasn't been ordered.

```python
pd.merge(df_products, df_order_details, on="product_id", how="left")
```

## Outer join of products and orders

_Which product id is missing in the products table?_
- Product id 5 is missing in the products table.

```python
pd.merge(df_products, df_order_details, on="product_id", how="outer")
```
